```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load dataset
df = pd.read_csv("/content/insurance (1).csv")
df.head()
```

```python
# Step 3: Basic info
df.info()
df.describe()
df.isnull().sum()
```

```python
# Step 4: Exploratory Data Analysis
sns.scatterplot(data=df, x='age', y='expenses')
plt.title("Age vs Expenses")

sns.scatterplot(data=df, x='bmi', y='expenses')
plt.title("BMI vs Expenses")

plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
# Step 5: Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['sex'] = le.fit_transform(df['sex'])
df['smoker'] = le.fit_transform(df['smoker'])
df['region'] = le.fit_transform(df['region'])
df.head()
```

```python
# Step 6: Split into Features and Target
X = df.drop('expenses', axis=1)
y = df['expenses']
```

```python
# Step 7: Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Step 8: Model training
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Step 9: Evaluation
print("Model Accuracy:", model.score(X_test, y_test))
```
